```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

img = plt.imread("images/highway.jpg")
print(img.shape)
plt.imshow(img)
plt.axis("off");
```

```python
gray_img = np.zeros(img.shape[:2])
print(gray_img.shape)
for i,row in enumerate(img):
    for j,pixel in enumerate(row):
        gray_img[i,j] = (0.299*pixel[0] + 0.587*pixel[1] + 0.114*pixel[2]) / 255
        
plt.imshow(gray_img,cmap = plt.get_cmap('gray'))
plt.axis("off");
print(gray_img[200:205,200:205])
```

```python
def gaussian(x,sigma):
    return np.exp(-np.square(x) / (2 * sigma**2) ) / sigma * np.sqrt(2 * np.pi)

x = np.arange(-3,3,0.01)
y = gaussian(x,sigma=0.3)
plt.plot(x,y);
```

```python
def gaussian_2d(X,Y,sigma):
    return np.exp(-(np.square(X) + np.square(Y)) / (2 * sigma**2) ) / sigma * np.sqrt(2 * np.pi)

x = np.arange(-3,3,0.01)
y = np.arange(-3,3,0.01)
X, Y = np.meshgrid(x, y)

Z = gaussian_2d(X,Y,sigma=0.5)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
def diff(x,y,differential):
    
    dy = np.zeros(len(y))
    delta_x = x[1] - x[0]
    
    for i in range(1,len(x)):
        dy[i] = (y[i] - y[i-1]) / delta_x
        
    dy[0] = dy[1]
    
    differential -= 1
    if differential <= 0:
        return dy
    else:
        return diff(x,dy,differential)

dy = diff(x,y,1)
plt.plot(x,dy);    
```

```python
d2y = diff(x,y,2)
plt.plot(x,d2y);
```
